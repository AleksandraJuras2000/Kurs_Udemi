### MNOŻENIE MACIERZY

Aby móc pomnożyć dwie macierze liczba kolumn w lewej macierzy musi zgadzać sie z liczba kolumn w prawej macierzy. Mnożenie macierzy nie jest przemienne

obraz_2024-12-06_114634513.png

```python
X = np.array([[5, 3], [3, 9]])
Y = np.array([[1], [-1]])
print(X)
print(Y)
```

```python
Z = np.dot(X, Y)
Z
```

```python
Z.shape
```

```python
Z = np.dot(Y, X)  #tu pojawi sie błąd
Z
```

### MACIERZ ODWROTNA
#### macierz kwadratowa stopnia n. Macierz A posiada macierz odwrotną gdy istnieje macierz B, taka że AB = BA = 1

```python
A
B = np.linalg.inv(A)
B
```

```python
np.set_printoptions(precision=4, suppress = True)
np.dot(A, B)
```

```python
np.dot(B, A)
```

### ROZWIĄZYWANIE UKŁADÓW RÓWNAŃ

```python
A = np.array([[2, 4], [1, -1]])
```

```python
B = np.array([[10], [-1]])
```

```python
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
X
```

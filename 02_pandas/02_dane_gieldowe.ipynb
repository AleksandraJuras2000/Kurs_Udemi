### OBLICZANIE NOWYCH KOLUMN

```python
df = fetch_financial_data('UBER')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Average'] = (df.Open + df.Close) / 2.0
df.head()
```

```python
df = df.sort_index()  # jesli nie przypiszemy to te zmiany ni zajda
df.head()
```

```python
df.Close.shift(1) #przesuwamy o jeden wiersz
```

```python
df['Close_shift'] = df.Close.shift(1)
df.head()
```

```python
df['Daily_Change'] = df.Close / df.Close_shift - 1 # dzienna zmiana wyrażona w procentach
df.head()
```

```python
# usuwanei kolumny
del df['Daily Change']
```

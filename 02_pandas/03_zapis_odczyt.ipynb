### WCZYTANIE DANYCH SPEK

```python
def fetch_financial_data(company = 'AMZN'):
  """
  This function fetch stock market quotations.
  """
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source='stooq')

apple = fetch_financial_data('AAPL')
amazon = fetch_financial_data('AMZN')
google = fetch_financial_data('GOOGL')
uber = fetch_financial_data('UBER')
```

```python
apple.info()
```

```python
uber.head()
```

### PREPROCESSING

```python
apple.columns = ['apple_' + col.lower() for col in apple.columns]
google.columns = ['google_' + col.lower() for col in google.columns]
uber.columns = ['uber_' + col.lower() for col in uber.columns]
amazon.columns = ['amazon_' + col.lower() for col in amazon.columns]
```

```python
apple.head()
```

```python
#poczenie danych czyli konkatenacja
df = pd.concat(objs=[apple, amazon, google, uber], axis=1)
```

```python
pd.set_option('display.float_format', lambda x: f'{x: .2f}')  # dwie liczby po przecinku
df.describe().T
```

```python
df.corr()
```

```python
df.columns
```

```python
closes = [col for col in df.columns if col.endswith('close')]  #list comprehesion
closes
```

```python
df[closes].corr()
```

```python
closes = [col for col in df.columns if col.endswith('close') or col.endswith('volume')]  #list comprehesion
closes
```

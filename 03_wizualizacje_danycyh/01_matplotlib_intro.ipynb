## MATPLOTLIB - INTRO
Matplotlib to popularna biblioteka programistyczna w języku Python, która umożliwia tworzenie wykresów i wizualizacji danych w sposób interaktywny lub statyczny. Jest to potężne narzędzie do prezentowania i analizy danych za pomocą różnego rodzaju grafik, wykresów i plotów.

Główne cechy i funkcje biblioteki Matplotlib:

### Wszechstronność:
Matplotlib umożliwia tworzenie różnorodnych typów wykresów, takich jak wykresy liniowe, wykresy słupkowe, wykresy punktowe, wykresy kołowe, histogramy, wykresy skrzynkowe (box plots) i wiele innych. Możesz dostosować wiele aspektów wyglądu i formatowania wykresu, aby dostosować go do swoich potrzeb.

### Intuicyjny interfejs:
Biblioteka oferuje intuicyjny i łatwy do zrozumienia interfejs do tworzenia wykresów. Programiści mogą kontrolować niemal każdy aspekt wyglądu wykresu, takie jak tytuł, etykiety osi, legenda, kolory i wiele innych, za pomocą prostych poleceń i funkcji.

### Wsparcie dla wielu formatów plików:
Matplotlib pozwala na zapisywanie wykresów w różnych formatach, takich jak PNG, JPEG, PDF, SVG i wiele innych. Możesz także osadzać wykresy w interaktywnych aplikacjach webowych lub używać ich w raportach i dokumentach.

### Interaktywność:
Choć Matplotlib jest często używany do tworzenia statycznych wykresów, to posiada również interaktywne narzędzia i funkcje, które pozwalają na interaktywną eksplorację danych w czasie rzeczywistym.

### Wsparcie dla wielu platform:
Matplotlib jest wieloplatformowy i działa na różnych systemach operacyjnych, co oznacza, że możesz tworzyć wykresy na dowolnej platformie, na której działa Python.

### Rozszerzalność:
Matplotlib jest bardzo rozszerzalny i można go łączyć z innymi bibliotekami do analizy danych, takimi jak NumPy, pandas, SciPy i inne, aby tworzyć bardziej zaawansowane wykresy i wizualizacje.

### Aktywna społeczność:
Matplotlib ma dużą i aktywną społeczność użytkowników i deweloperów, co oznacza, że istnieje wiele dostępnych źródeł, tutoriali i pomocy online.



Matplotlib jest niezastąpionym narzędziem w dziedzinie wizualizacji danych i jest szeroko stosowany w dziedzinach takich jak nauka danych, analiza biznesowa, badania naukowe, edukacja i wiele innych. Dzięki swojej wszechstronności i elastyczności umożliwia programistom i analitykom danych tworzenie czytelnych i atrakcyjnych wykresów do prezentacji i analizy danych.

```python
import matplotlib
matplotlib.__version__
```

```python
import numpy as  np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set()  # tutaj ustawiamy motyw domyślny
```

```python
fig = plt.figure() # tu rozpoczynamy zabawe z figura, przypisujemy do zmiennej
fig.show() # tu pokazujemy
```

```python
_ = plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1], [0, 3], label='line', color = 'green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color = 'green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos) # narzucamy kolejne warstwy na obiekt typu figure!!!!!!!!!!!!!! nie zamieniamy
```

## Tworzenie wielu wykresów

```python
plt.subplot(121)  # 1- podajemy liczbe wierszy ktore chcamy uzyskać, 2- liczb akolumn które chcemy uzyskać, 1- odnosi sie do konkretnego wykresu
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 3, 2], [3, 4, 5], 'r--')
```

```python
fig = plt.figure(figsize=(12, 8))  # w calach wielkość


plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 3, 2], [3, 4, 5], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 2, 4], 'ro')

fig.show() #w notebookach zostaje domyślnie pokazany
```

## Wykresy słupkowe

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])  # HHH  BARH  --> bo horyzontalnie
```

## Wykresy punktowe

```python
x1 = np.random.randn(300)
x2 = np.random.rand(300)

plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.rand(300)*50
plt.scatter(x=x1, y=x2, s=x3)  # pararmetr s odpowiada za rozmiar danych
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha = 0.5) # parametr alpha odpowiada za przezroczystość
```

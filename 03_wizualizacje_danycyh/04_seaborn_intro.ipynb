## Seaborn - Intro

Seaborn to biblioteka programistyczna w języku Python, która jest używana do tworzenia atrakcyjnych i informatywnych wykresów statystycznych. Jest to oparta na Matplotlib biblioteka, która dostarcza wyższy poziom abstrakcji i łatwiejszy interfejs do tworzenia wykresów, szczególnie tych związanych z analizą danych i statystyką.



Główne cechy i funkcje biblioteki Seaborn:

### Wygodna integracja z danymi:
Seaborn jest zintegrowany z biblioteką Pandas, co oznacza, że można łatwo używać struktur danych DataFrame w Pandas do tworzenia wykresów. To ułatwia manipulację danymi i generowanie wykresów na podstawie istniejących zbiorów danych.

### Wykresy statystyczne:
Seaborn skupia się na wykresach statystycznych, takich jak wykresy punktowe, wykresy pudełkowe (box plots), wykresy rozkładu, wykresy wiolinowe, wykresy słupkowe itp. Dzięki tym wykresom można wizualizować rozkłady danych i relacje między nimi.

### Ustawienia domyślne estetyki:
Seaborn dostarcza przyjemne domyślne style i kolory do wykresów, co pozwala na tworzenie atrakcyjnych i czytelnych wykresów bez konieczności wielu ręcznych dostosowań.

### Wsparcie dla kolorów i palet:
Biblioteka oferuje szeroką gamę kolorów i palet kolorów, które można dostosować do konkretnych potrzeb wizualizacji danych.

### Dodatkowe funkcje do wizualizacji:
Seaborn posiada również dodatkowe funkcje do wizualizacji, takie jak funkcje do tworzenia macierzy korelacji, wykresy regresji, wykresy heatmap i wiele innych.

### Łatwa personalizacja:
Choć Seaborn dostarcza domyślnych ustawień, umożliwia także pełną personalizację wykresów, co pozwala dostosować je do konkretnych potrzeb projektu.

### Wysoka jakość wyjścia:
Wykresy generowane przez Seaborn są zazwyczaj w wysokiej jakości i gotowe do użycia w raportach, prezentacjach lub publikacjach.

### Aktywna społeczność:
Seaborn ma aktywną społeczność użytkowników i deweloperów, co oznacza, że istnieje wiele dostępnych źródeł, tutoriali i pomocy online.



Seaborn jest często wybierany przez analityków danych i naukowców danych do tworzenia wykresów statystycznych i wizualizacji danych, zwłaszcza w kontekście analizy eksploracyjnej. Dzięki swojej prostocie i estetyce pomaga w szybkim zrozumieniu danych i komunikowaniu wyników analizy.



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
sns.__version__
```

## Tips data

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip')
```

```python
sns.set(font_scale=1.1)
_ = sns.relplot(data=df, x='total_bill', y='tip', hue='time') # trzecia zmienna  kolor
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size') #trzecia zmienna rozmiar
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue = 'size')
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue ='size', col='time') #rozbije nam na dwa wykresy wzgledem zmiennej time
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue ='size', row = 'time')
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue ='size', col='time', row = 'smoker')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind = 'swarm')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind = 'box')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind = 'violin')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind = 'bar')
```

```python
_ = sns.catplot(data=df, x='day', kind = 'count')  # wykres częstotliwości
```

```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
import seaborn as sns

sns.set()
```

```python
np.random.seed(10)
np.set_printoptions(precision=6, suppress=True)

def sigmoid(x):
  return 1/(1+np.exp(-x))

X = np.arange(-5, 5, 0.1)
y = sigmoid(X)
plt.plot(X, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')
# funckja spłaszcza nasze wawrtości do przedzialu 0-1
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
data
```

```python
# najpierw z tego naszego data wyodrebnimy dane i zminna docelową
X = data.data
y = data.target

print(f'shape X: {X.shape}')
print(f'shape y: {y.shape}')
```

```python
df = pd.DataFrame(X, columns=data.feature_names)
df.info()
```

## PODZIAŁ DANYCH NA ZBIÓR TRENINGOWY I TESTOWY

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)  # nie ustaliliśmy tutaj test size a domyślnie jest on rózny 25% czyli 1/4 zbioru idzie na testowy

print(f'shape X_train: {X_train.shape}')
print(f'shape X_test: {X_test.shape}')
print(f'shape y_train: {y_train.shape}')
print(f'shape y_test: {y_test.shape}')
```

## DOPASOWANIE MODELU

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

## PREDYKSJA NA PODTSAWIE MODELU

```python
y_prob = classifier.predict_proba(X_test)
y_prob  # zwraca dla każdej probki prawdopodobieństwo do trafdienia do klasy 1 i 0 w wierszach sumuje sie do 1
# potem zwraca nam wynik poprzez wziecie wiekszego prawdopodobieństwa
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

## OCENA MODELU

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
import seaborn as sns
sns.set()

cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm)

acc = accuracy_score(y_test, y_pred)
print(f'accuracy: {acc}')
```

```python
def plot_confusion_matrix(cm):
  cm = cm[::-1]
  cm = pd.DataFrame(cm, columns=['pred_0', 'pred_1'], index=['true_0', 'true_1'])

  fig = ff.create_annotated_heatmap(z=cm.values, x=list(cm.columns), y=list(cm.index),
                                    colorscale='ice', showscale=True, reversescale=True)
  fig.update_layout(width=500, height=500, title_text=f'confusion matrix, Accuracy: {acc:.4f}', font_size=16)
  fig.show()

plot_confusion_matrix(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred, target_names=data.target_names))
#nowotwór łagodny jest wykrywany z wyższą precyzją (benign -> 0.94)
```

```python
# typowy model do klasyfikacji
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
sns.set()
```

```python
from sklearn.datasets import load_iris
raw_data = load_iris()
raw_data.data
```

```python
raw_data.target
```

```python
df1 = pd.DataFrame(raw_data.data, columns= raw_data.feature_names)
df2 = pd.DataFrame(raw_data.target, columns= ['class'])
df = pd.concat([df1, df2], axis=1)
df
```

```python
X = raw_data.data
y = raw_data.target
X=X[:, :2]  # wybieramy tutaj tylko dwie zmienne które nas interesuja do tego modelu

print('X shape:', X.shape)
print('y shape:', y.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Wykres punktowy')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
df = pd.DataFrame(X, columns=['sepal_length', 'sepal_width'])
target = pd.DataFrame(y, columns=['class'])
df = pd.concat([df, target], axis=1)
px.scatter(df, x='sepal_length', y='sepal_width', color='class', width=600, height=400)
```

## K-nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)  # tutaj liczba sąsiadów na których bedzimy patrzeć
classifier.fit(X, y)
# klasa która przezważa zostaje wybrana do naszego punktu
```

```python
accuracy = classifier.score(X, y)
print('Accuracy:', accuracy)
```

## WYKRES GRANIC DECYZYJNYCH

```python
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
mesh = np.c_[xx.ravel(), yy.ravel()]
Z = classifier.predict(mesh)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(9, 7))
plt.pcolormesh(xx, yy, Z, cmap='gnuplot', alpha=0.1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='gnuplot', edgecolors='r')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title(f'3-class classification k=5, accuracy: {accuracy:.4f}')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
for i in range(1, 7):
    plt.subplot(3, 2, i)
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier.fit(X, y)
    accuracy = classifier.score(X, y)
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    mesh = np.c_[xx.ravel(), yy.ravel()]
    Z = classifier.predict(mesh)
    Z = Z.reshape(xx.shape)
    plt.pcolormesh(xx, yy, Z, cmap='gnuplot', alpha=0.5, colorizer='red')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='gnuplot', edgecolors='r')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f'3-class classification k={i}, accuracy: {accuracy:.4f}')
plt.show()
```

## GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV

grid_params = {'n_neighbors': range(2, 30)}

classifier = KNeighborsClassifier()

gs = GridSearchCV(classifier, grid_params, cv=3)  #kross walidacja na trzech foldach.
#dzieli dane na 3 równe częsci i łapie dwie pierwzse czesci i trenuje a sprawdza na ostatniej,
#pozniej trenuje na drugiej i trzeciej i sprawdza na pierwszej i tak dalej..
gs.fit(X, y)
```

```python
gs.best_params_ # jaka liczba sąsiadów jest najlepsza dla naszego modelu
```

```python
k = gs.best_params_['n_neighbors']
k
```

```python
classifier = gs.best_estimator_ # najlepszy estymator czyli ten model z nasza liczba sąsiadów równą 6
classifier
```

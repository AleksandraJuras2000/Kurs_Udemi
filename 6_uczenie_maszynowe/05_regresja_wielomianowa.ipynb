```python
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
```

image.png

```python
np.random.seed(20)
X = np.arange(-10, 10, 0.5)
noise = 80 * np.random.rand(40)
y = -X**3 + 10*X**2 - 2*X + 3 + noise
X = X.reshape(40, 1)
_=plt.scatter(X, y)
```

```python
# spróubjemy najpierw dopasowac model liniowy i zobaczymy jaki bedzie bład tego modelu
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
y_pred = regressor.predict(X)

plt.scatter(X, y)
_ = plt.plot(X, y_pred, color='red')


# no widzimy że nie jest to dobre rozwiązanie
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)  # to nie jest dobry wynik
```

## EKSTRAKCJA CECH WIELOMIANOWYCH - Stopien 2

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)   # jkai stopieńwielomianu
X_poly = poly.fit_transform(X)
X_poly  # 1 to nasza stała, w drugiej kolumnie to nasza zmienna a w trzeciej to zmienna podniesiona do kwadratu
```

## DOPASOWANIE MODELU

```python
regressor_poly = LinearRegression()
regressor_poly.fit(X_poly, y)

y_pred = regressor_poly.predict(X_poly)

plt.scatter(X, y, color = 'blue')
_ = plt.plot(X, y_pred, color = 'red')

#tym razem jest dużo lepiej
```

```python
r2_score(y, y_pred) # leeepiej
```

## EKSTRAKCJA CECH WIELOMIANOWYCH - STOPIEŃ 3

```python
np.set_printoptions(suppress=True)  # tł tłumienie

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
X_poly   # w ostatniej kolumnie jest nasza wartosc(kol 2) podniesiona do pot 3
```

```python
regressor_poly_3 = LinearRegression()
regressor_poly_3.fit(X_poly, y)

y_pred = regressor_poly_3.predict(X_poly)

plt.scatter(X, y, color = 'blue')
_ = plt.plot(X, y_pred, color = 'red')
```

```python
r2_score(y, y_pred)
```
